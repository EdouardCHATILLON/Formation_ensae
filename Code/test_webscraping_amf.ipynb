```python
#Selon chat gpt il vaut mieux utiliser l'API interne du site
import requests

def get_vl(part_id):
    url = f"https://geco.amf-france.org/api/produits/part/{part_id}"
    
    response = requests.get(url, headers={"User-Agent": "CepeBot"})
    response.raise_for_status()
    
    data = response.json()
    
    return {
        "vl": data.get("valeurLiquidative"),
        "date_vl": data.get("dateValeurLiquidative")
    }

result = get_vl(1487)

print(result)
```
